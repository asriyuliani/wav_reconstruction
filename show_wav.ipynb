```python
import librosa 
import IPython.display
import numpy as np

# basic constants
SR = 8000
N_FFT = 256
HOP_SIZE = 128
```

```python
# Utility for display / audio playback 
# from https://nbviewer.jupyter.org/gist/naotokui/70eccf60e703f8cc601d79a17c8a7d87/phase_reconstruction_test.ipynb

import matplotlib.pyplot as plt
import librosa.display

def show_audio_array(y_r):
    # write
    filename_tmp = "_tmp.wav"
    librosa.output.write_wav(filename_tmp, y_r, sr=SR)
    IPython.display.display(IPython.display.Audio(filename_tmp))
    
def show_waveplot(y, sr):
    plt.figure()
    plt.subplot(3, 1, 1)
    librosa.display.waveplot(y, sr=sr)
    plt.title('Waveform')
    plt.show()
    
def show_melspec(spec):
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(librosa.power_to_db(spec, ref=np.max),
                          y_axis='mel', fmax=4000,
                          x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')
    plt.tight_layout()
    plt.show()

def show_melspec_audio(y, melspec_bands):
    melspec = librosa.feature.melspectrogram(y=y, sr=SR,
                n_fft=N_FFT, hop_length=HOP_SIZE, n_mels=melspec_bands)
    show_melspec(melspec)
```

```python
# Original waveform
filename = "FAK_1237743A.wav"
filename_tmp = "_tmp.wav"

# test data 
y, _ = librosa.load(filename, sr=SR, mono=True)
show_audio_array(y)
show_waveplot(y, SR)
show_melspec_audio(y, 128)
```
